```python
def write_parquet(year, color, schema):
    
    for month in range(1, 13):
        print(f"Processing data for {year}/{month}...\n")
        
        input_path = f"data/raw/{color}/{year}/{month:02d}"
        output_path = f"data/pq/{color}/{year}/{month:02d}"
        
        df_spark = spark.read \
            .option("header", "true") \
            .schema(schema) \
            .csv(input_path)
        
        df_spark \
            .repartition(4) \
            .write.parquet(output_path)
```

```python
write_parquet(2021, 'green', green_schema)
```

```python
write_parquet(2020, 'green', green_schema)
```

```python
write_parquet(2021, 'yellow', yellow_schema)
```
